# we can put our attributes as positions


```python
nx.get_node_attributes(O , "role")
```

```python
node_color = [O.degree(v) for v in O]
node_color
```

```python
node_size = [0.0005 * nx.get_node_attributes(O , "role")[v] for v in O]
node_size
```

```python
plt.figure(figsize=(8,6))
pos = nx.circular_layout(O)
nx.draw_networkx(O ,pos, node_size= node_size , node_color= node_color,
                alpha = 0.7 , with_labels = False  , cmap = plt.cm.PuRd)
plt.axis("off")
plt.tight_layout()
```
